### Women and children first?

*** 1. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (25 pts)*** 

```python
t_df.columns = t_df.iloc[0]
t_df.drop(t_df.index[0], inplace=True)
t_df.head()
```

Shows How many total passengers there are for each gender.

```python
survival = t_df.copy()
survival.groupby('sex')['survived'].count()
```

Shows the number of how many people of each gender had died and survived.

```python
sex_ratio = t_df.copy()
sex_ratio['proportion'] = 1
sex_ratio = sex_ratio.groupby(['sex', 'survived']).agg({'proportion': 'sum'})
print(sex_ratio)
```

Calulates and prints the proportion of each gender that survived and died.

```python
survivedProportion = sex_ratio.groupby(level=0).apply(lambda x:100 * x/x.sum())
print(survivedProportion)
```

## This calculates and prints the percentage of each gender that had survived. Using a try and except block.

```python
part1= t_df.copy()
sex_survive = part1.groupby(['sex','survived'])
sex = ['female', 'male']

for gender in sex:
    try:
        death = len(sex_survive.get_group((gender, 0)))
        survived = len(sex_survive.get_group((gender, 1)))
        print('\n The percentage of survival for', gender,': ', (survived/(death+survived)*100))
    except:
        continue
```

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. ( pts)***

#### This makes a new table and lables each age category accordingly.

```python
new_t_df = t_df.copy()

new_t_df["Age_Group"] = ""
for x in range(1, len(new_t_df)+1):
    if (new_t_df.age[x] < 14.0): 
        new_t_df.Age_Group[x] = "children"
    elif (new_t_df.age[x] > 13.0) and (new_t_df.age[x] < 21.0):
        new_t_df.Age_Group[x] = "adolescents"
    elif (new_t_df.age[x] > 20.0) and (new_t_df.age[x] < 65.0):
        new_t_df.Age_Group[x] = "adult"
    elif (new_t_df.age[x] > 64.0):
        new_t_df.Age_Group[x] = "senior"
    else:
        new_t_df.Age_Group[x] = "NaN"
new_t_df.head(20)        
```

This is the proportions of how many survived and died in each age, sex, and class category.

```python
new_t_df['proportion'] = 1
age_group = new_t_df.groupby(['Age_Group','pclass','sex', 'survived']).agg({'proportion': 'sum'})
print(age_group)
```
