## Plotting Titanic Dataset

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.io as pio
pio.kaleido.scope.default_format = "png"
from IPython.display import Image
import plotly.express as px
import plotly.figure_factory as ff
```

```python
titanic = pd.read_excel("../data/titanic.xls", "titanic")
titanic.head()
```

#### 1. Create a bar chart for number of passengers survived by class

##### This is creating a bar graph to have the amount of passengers that survived by class.

```python
fig = px.bar(titanic, x='pclass', y='survived')
bar_result = Image(fig.to_image(format="png", height=1000, width=1000))
bar_result
```

#### 2. Create a vertical bar chart for passengers survived by class and gender

#####  This is creating a bar graph that shows the amount of passsengers that had survived by class and gender. 

```python
fig = px.bar(titanic, x='pclass', y='survived', color='sex', barmode='group')
vertical_bar_result = Image(fig.to_image(format="png", height=1000, width=1000))
vertical_bar_result
```

#### 3. Create a stacked chart survived vs dead for different classes and gender

##### This is a stacked histogram of the passengers that had survived and died by class and gender. The color repesents who survived and died.

```python
#fig = px.histogram(titanic, x='survived', color='sex', facet_col='pclass', barmode='stack')
fig = px.histogram(titanic, x='sex', color='survived', facet_col='pclass', barmode='stack')
stacked_result = Image(fig.to_image(format="png", height=1000, width=1000))
stacked_result
```

#### 4. Plot histogram of fares paid for passengers bins = 30

##### This is a histogram of the fares that were paid by the passengers. It has a bin of 30.

```python
fig = px.histogram(titanic, x="fare", nbins=30)
histogram_result = Image(fig.to_image(format="png", height=1000, width=1000))
histogram_result
```

#### 5.Create a KDE vs bar plot of the fare

##### This is an KDE with a rug displaying the fare. 

```python
x = np.random.randn(1000)
group_labels = ['fare']
fig = ff.create_distplot([x],group_labels,curve_type='kde')
dist_result = Image(fig.to_image(format="png", height=1000, width=1000))
dist_result
```

##### This is a bar plot of the fare instead of the histogram from the last question.

```python
fig = px.bar(titanic, y="fare")
fare_bar_result = Image(fig.to_image(format="png", height=1000, width=1000))
fare_bar_result
```

#### 6. Create a box plot of the fares by passenger class

##### This makes a box plot of the titanic fare amount for each class.

```python
fig = px.box(titanic, y="fare",x='pclass')
box_result = Image(fig.to_image(format="png", height=1000, width=1000))
box_result
```

#### 7. Create a box plot of the age by passenger class

##### This is a box plot to relate the age to the passengers of each class.

```python
fig = px.box(titanic, y="age", x='pclass')
age_box_result = Image(fig.to_image(format="png", height=1000, width=1000))
age_box_result
```
